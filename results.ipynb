Here each experiment is run only once. So the results may not be accurate enough.

````python
_, ax = plt.subplots(figsize=(15, 15))

ax.errorbar(x=nchains, y=jrevenues, fmt='r--', label='joint', yerr=jrevenues_err)
ax.errorbar(x=nchains, y=drevenues, fmt='b--', label='disjoint', yerr=drevenues_err)

ax.set_title('FatTree')
ax.set(ylabel='Revenue', xlabel='Chains')
ax.legend()
````

## Topology: USnet

````python
# result is a map between number of the chains and costs of their placement
# jresult is a result of joint problem and dresult is a result of disjoint problem

jresult = {
    50: 23500,
    60: 27500,
    70: 31400,
    80: 33500,
    90: 35800,
    100: 38700,
    130: 38900,
}

dresult = {
    50: 22500,
    60: 26000,
    70: 30000,
    80: 32000,
    90: 34400,
    100: 37500,
    130: 37800,
    
}

_, ax = plt.subplots(figsize=(15, 15))

chains = [d for d in jresult]
jcost = [jresult[d] for d in jresult]
dcost = [dresult[d] for d in dresult]

ax.plot(chains, jcost, 'r--', label='joint')
ax.plot(chains, dcost, 'b', label='disjoint')
ax.set_title('USNet')
ax.set(ylabel='Cost', xlabel='Chains')
ax.legend()
````
